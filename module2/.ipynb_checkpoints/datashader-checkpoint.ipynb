For module 2 we'll be looking at techniques for dealing with big data. In particular binning strategies, and the datashader library (which probably proves we'll never need to bin large data for visualization ever again.)

To demonstrate these concepts we'll be looking at the PLUTO dataset put out by New York City's department of city planning. PLUTO contains data about every tax lot in New York City.

PLUTO data can be downloaded from [here](https://www1.nyc.gov/site/planning/data-maps/open-data/dwn-pluto-mappluto.page). Unzip them to the same directory as this notebook, and you should be able to read them in using this (or very similar) code. Also take note of the data dictionary, it'll come in handy for this assignment.

```python
bk = pd.read_csv('PLUTO17v1.1/BK2017V11.csv')
bx = pd.read_csv('PLUTO17v1.1/BX2017V11.csv')
mn = pd.read_csv('PLUTO17v1.1/MN2017V11.csv')
qn = pd.read_csv('PLUTO17v1.1/QN2017V11.csv')
si = pd.read_csv('PLUTO17v1.1/SI2017V11.csv')

ny = pd.concat([bk, bx, mn, qn, si], ignore_index=True)

# Getting rid of some outliers
ny = ny[(ny['YearBuilt'] > 1850) & (ny['NumFloors'] != 0)]

#Defining some helper functions for DataShader
background = "black"
export = partial(export_image, background = background, export_path="export")
cm = partial(colormap_select, reverse=(background!="black"))
```

I'll also be bringing in some geo data for this dataset. 

All boros

```python
featurelist = [] 
boros = ['qn', 'bk', 'si', 'bx', 'mn']

for boro in boros:
    apiurl = 'https://pluto.cartodb.com/api/v2/sql?q=SELECT%20*%20FROM%20'+boro+'_mappluto_13v1&format=GeoJSON'
    with urllib.request.urlopen(apiurl) as url:
        geodata = json.loads(url.read().decode())
        featurelist.extend(geodata['features'])
        
lats = []
lons = []
bbls = []
count = 0
for feature in featurelist:
    try:
        points = np.array(shape(feature['geometry']).representative_point())
        bbls.append(feature['properties']['bbl'])
        lons.append(points[0])
        lats.append(points[1])
    except:
        count=count+1
        continue

geos = pd.DataFrame(dict(lat = lats, lon = lons, BBL = bbls))

merged = pd.merge(ny, geos, how='inner', on='BBL')
```

Manhattan

```python
siteurl = 'https://pluto.cartodb.com/api/v2/sql?q=SELECT%20*%20FROM%20mn_mappluto_13v1&format=GeoJSON'

with urllib.request.urlopen(siteurl) as url:
    geodata = json.loads(url.read().decode())

lats = []
lons = []
bbls = []
count = 0
for feature in geodata['features']:
    try:
        points = np.array(shape(feature['geometry']).representative_point())
        bbls.append(feature['properties']['bbl'])
        lons.append(points[0])
        lats.append(points[1])
    except:
        count=count+1
        continue
    
geos = pd.DataFrame(dict(lat = lats, lon = lons, BBL = bbls))

manhattan_merged = pd.merge(ny, geos, how='inner', on='BBL')
```

Mapping out points

```python
manhattan_merged['NumFloorsCat'] =\
    pd.Categorical((np.floor(manhattan_merged['NumFloors']/10)*10).apply(lambda x: int(x)))
```

```python
NewYorkCity   = (( -74.05,  -73.85), (40.65, 40.91))
cvs = ds.Canvas(500, 800, *NewYorkCity)
agg = cvs.points(manhattan_merged, 'lon', 'lat')

tf.shade(agg, cmap=colormap_select(viridis), how='log')
```

Assessed value land vs assessed value building

First lets analyze these variables

```python
ny.loc[ny['AssessLand'], 'AssessLand']
```

```python
al33 = np.percentile(manhattan_merged['AssessLand'].values, 33)
al66 = np.percentile(manhattan_merged['AssessLand'].values, 66)
at33 = np.percentile(manhattan_merged['AssessTot'].values, 33)
at66 = np.percentile(manhattan_merged['AssessTot'].values, 66)

manhattan_merged['AssessLand3tile'] = manhattan_merged['AssessLand']\
                                         .apply(lambda x: 1 if x < al33 else 2 if x < al66 else 3)
manhattan_merged['AssessTot3tile'] = manhattan_merged['AssessTot']\
                                         .apply(lambda x: 1 if x < at33 else 2 if x < al66 else 3)

manhattan_merged['AssessAll3tile'] =\
                     manhattan_merged['AssessLand3tile'].apply(str) + manhattan_merged['AssessTot3tile'].apply(str) + 'c'

manhattan_merged['AssessAll3tile'] = pd.Categorical(manhattan_merged['AssessAll3tile'])
    
color_key = {
    '11c': '#e8e8e8',
    '12c': '#b0d5df',
    '13c': '#64acbe',
    '21c': '#e4acac',
    '22c': '#ad9ea5',
    '23c': '#627f8c',
    '31c': '#c85a5a',
    '32c': '#985356',
    '33c': '#574249'
}

# greener is Assessed Land, redder is Assessed Total, Maroon is both, 
        
```
